```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(np.arange(0.001,11,0.001))
# y = 10+(np.log(x/10)/.5)
y = 10*np.exp(.5*(x-10))

fig, ax = plt.subplots()
# ax.set_title('LOTR')
ax.set_ylabel('LOTR Limit Token Value')
ax.set_yticks([2,4,6,8], minor=True)
ax.set_yticks([0,10], minor=False)
ax.set_yticklabels([0,'Limit Price'])
ax.yaxis.grid(True, which='major')
ax.set_ylim([0,10])
ax.set_xticks([2,4,6,8], minor=True)
ax.set_xticks([0,10], minor=False)
ax.set_xticklabels([0,'Limit Price'])
ax.set_xlabel('Target Asset Market Price')
ax.xaxis.grid(True, which='major')
ax.set_xlim([0,10])
ax.plot(x, y, color='purple', marker='', linestyle='solid', linewidth=3, markersize=0)
# plt.draw()
plt.savefig('./litepaper/images/graph_limit_below.png', dpi=100, bbox_inches='tight')
plt.show()
```
